```python
import pandas as pd
import pandas as pd
import json
import numpy as np
import requests as r
paris = pd.read_csv('listings.csv')
#Checking out the data
paris.head(100)
```

```python
#Noticed was that the neighbourhood_group column contained a lot of nulls
paris['neighbourhood_group'].isnull().value_counts()
```

```python
#Found it unnecessary
paris = paris.drop(['neighbourhood_group'],axis=1)
```

```python
#Checking how many values are in each column and how they're represented to clean them
print(paris.isnull().sum(),'\n')
paris['host_name'] = paris['host_name'].fillna('N/A')
paris['name'] = paris['name'].fillna('N/A')
paris['last_review'] = paris['last_review'].fillna('N/A')
paris['reviews_per_month'] = paris['reviews_per_month'].fillna(0)
print(paris.isnull().sum())
# paris.to_json(orient='index')
```
